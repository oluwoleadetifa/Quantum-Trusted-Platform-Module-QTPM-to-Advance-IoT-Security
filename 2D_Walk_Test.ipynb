```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
QRNG_NO = pd.read_csv('QRNG.txt', header=None, sep=',', engine='python')
```

```python
QRNG_NO = QRNG_NO.T.reset_index(drop=True)
```

```python
#Data frames
Mers_Twist = pd.read_csv('mersenne_twist.txt', header=None, sep='\s+', engine='python')
Xor_Shift = pd.read_csv('xorshiftt.txt',  header=None, sep='\s+', engine='python')
```

```python
len(Xor_Shift)
```

```python
def random_2d_walk(arr:list)->list:

  # Select 5000 random indices from the dataset
  random_indices = np.random.choice(len(arr), size=5000, replace=False)

  # Initialize starting position for the random walk
  current_position = [0, 0]
  position = []

  # Perform 2D random walk
  for idx in random_indices:
      # Generate random increments of -1, 0, or 1 for each step
      increment_x = np.random.choice([-1, 0, 1])
      increment_y = np.random.choice([-1, 0, 1])

      # Update current position
      current_position[0] += increment_x
      current_position[1] += increment_y

      position.append([current_position[0], current_position[1]])

  return position
```

```python
#  XORSHIFT 2D RANDOM WALK TEST

position_Xor = random_2d_walk(Xor_Shift)
```

```python
#XORSHIFT PLOT

# Extract x and y coordinates from position
x_coords, y_coords = zip(*position_Xor)

# Plot the random walk
plt.plot(x_coords, y_coords)
plt.title("2D Random Walk For XORShift PRNG")
plt.xlabel("X-Coordinate")
plt.ylabel("Y-Coordinate")
plt.xlim(left=-60, right=20)
plt.ylim(bottom=-50, top=30)
plt.show()
```

```python
#  Mersenne twister 2D RANDOM WALK TEST

position_Mers = random_2d_walk(Mers_Twist)
```

```python
#MERSENNE PLOT

# Extract x and y coordinates from position
x_coords, y_coords = zip(*position_Mers)

# Plot the random walk
plt.plot(x_coords, y_coords)
plt.title("2D Random Walk Mersenne twister PRNG")
plt.xlabel("X-Coordinate")
plt.ylabel("Y-Coordinate")
plt.xlim(left=-60, right=20)
plt.ylim(bottom=-50, top=30)
plt.show()
```

```python
#  QRNG 2D RANDOM WALK TEST

position_QRNG = random_2d_walk(QRNG_NO)
```
