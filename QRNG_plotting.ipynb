```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
data_sizes = [1000, 10000, 100000, 1000000]
qrng = [7.589, 76.074, 764.8392, 7648.3337]
mers_twist = [0.0004, 0.0023, 0.0087, 0.1322]
xorshift = [0.0005, 0.0027, 0.0054, 0.132]

# Create the plot
plt.figure(figsize=(10, 6))

# Plot each RNG
plt.plot(data_sizes, qrng, marker='o', label='QRNG')
plt.plot(data_sizes, mers_twist, marker='s', label='Mersenne Twister')
plt.plot(data_sizes, xorshift, marker='^', label='XorShift')
plt.rcParams.update({'font.size': 20})

# Set x-axis to logarithmic scale
plt.xscale('log')

# Set y-axis to logarithmic scale
plt.yscale('log')

# Add labels and title
plt.xlabel('Data Size', fontsize = 28)
plt.ylabel('Time (seconds)', fontsize = 28)
plt.title('RNG Time Performance Comparison', fontsize = 35)

# Add legend
plt.legend()

# Add grid for better readability
plt.grid(True, which="both", ls="-", alpha=0.2)

plt.xticks(fontsize=25)
plt.yticks(fontsize=25)

# Show the plot
plt.tight_layout()
plt.show()
```
